```python
import pandas as pd
```

```python
enrol = pd.read_csv("enrolment-by-institutions.csv")
enrol.head()
```

```python
enrol.dtypes #notice that some of the cols that are supposed to be all numbers are object type instead of int64 type. Later found out mistake in data entry '-' instead of NaN
```

```python
enrol.replace('-', '0', inplace=True)
```

```python
enrol = enrol.astype({'smu':'int64',
              'sit':'int64',
              'sutd':'int64',
              'suss':'int64',
              'temasek_polytechnic':'int64',
              'nanyang_polytechnic':'int64',
              'republic_polytechnic':'int64',
              'lasalle_diploma':'int64',
              'lasalle_degree':'int64',
              'nafa_diploma':'int64',
              'nafa_degree':'int64'},
             )
```

```python
enrol.dtypes
```

```python
enrol.head()
```

```python
enrol.tail()
```

```python
enrol.shape #there are 17 different schools
```

```python
nus = enrol[["year", "sex", "nus"]]
nus.head()
```

```python
nusFemale = nus[(nus.sex=="F")]
nusFemale.reset_index(inplace=True)
```

```python
nusAll = nus[(nus.sex=="MF")] #all sex
nusAll.reset_index(inplace=True) #the reset_index function automatically creates a new col starting from the left, take note of this!
```

```python
nusMaleNumbers = nusAll["nus"]-nusFemale["nus"]
nusMale = pd.concat([nusFemale["year"], nusMaleNumbers], axis=1)
```

```python
nusMale.head()
```

Here comes the visualizations!

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.plot(nusAll["year"], nusAll["nus"], label="all")
ax.plot(nusFemale["year"], nusFemale["nus"], label="F")
ax.plot(nusMale["year"], nusMale["nus"], label="M")
ax.legend()
plt.show()
```

```python
enrol.head()
```

```python
fig, ax = plt.subplots(17, 1, figsize=(10,75))
for r in range(17):
  currSch = enrol.iloc[:,[0, 1, (r+2)]] #3 cols: year, sex, schName (numbers) #start from nus
  currAll = currSch[currSch.sex == "MF"]
  currAll.reset_index(inplace=True) #after reset_index, now has 4 cols: previous_index, year, sex, schName
  currFemale = currSch[(currSch.sex=="F")]
  currFemale.reset_index(inplace=True) #after reset_index, now has 4 cols: previous_index, year, sex, schName
  currMaleNumbers = currAll.iloc[:,3]-currFemale.iloc[:,3]
  currMale = pd.concat([currAll.iloc[:,[0,1,2]], currMaleNumbers], axis=1)
  currList = [currAll, currFemale, currMale]
  currListName = ["all","F","M"]
  for c in range(3):
    ax[r].plot(currList[c].iloc[:,1], currList[c].iloc[:,3], label=currListName[c])
    ax[r].set_title(currSch.columns[2])
    ax[r].legend()
plt.show()
```
